# 함수형 프로그래밍
- 여러 개의 함수를 작성해서 함수에 기반해 프로그램이 실행
- 코드의 재사용 가능
- 유지보수가 쉬워짐
- 버그 발생률이 낮고 예측 가능성이 높음
- 병렬처리, 동시성 처리에 강함 - 멀티 코어를 이용할 때 유리

```python
# 리스트의 제곱합 구하기
nums = [1,2,3,4]
squares = list(map(lambda x: x ** 2,  nums))
print(squares)
```

- 클래스 기반으로 프로그래밍

- 여러사람이 이용할 때 같은 코드를 독립적으로 사용 가능

```python
class NumberList:
    def __init__(self, numbers):
        self.numbers = numbers
    def squared(self):
        return [x**2 for x in self.numbers]

squared = NumberList([1, 2, 3, 4]).squared()
```

# 클래스가 필요한 이유
- 이전에 계산했던 결과를 기억하는 계산기

```python
result_cal =0
def add(num, result_cal):
    result_cal += num
    print("함수 내부의 result_cal id: ", id(result_cal))
    return result_cal
# 함수 내부의 매개변수는 함수 내부에서만 사용됨
```

```python
add(3, 0)
print("함수 내부의 result_cal id: ", id(result_cal))
```

```python
print(result_cal)
```

- 전염변수를 함수 내부에서만 사용할 수 있도록 하는 방법
    - global 전역변수명

```python
result_cal =0
def add(num):
    global result_cal 
    result_cal += num
    print("함수 내부의 result_cal id: ", id(result_cal))
    return result_cal
```

```python
print(add(1))
print("함수 외부의 result_cal id: ", id(result_cal))
```

```python
class FourCal():
    def setdata(self, num1, num2): 
        self.num1 = num1
        self.num2 = num2
        print("self.num1", self.num1)
        print("self.num2", self.num2)
        
    def add(self):
        result = self.num1 + self.num2
        return result
    
    def sub(self):
        result = self.num1 - self.num2
        return result
```

```python
class FourCal2():
    def __init__(self, num1, num2):
        self.num1 = num1
        self.num2 = num2
        print("self.num1", self.num1)
        print("self.num2", self.num2)
        
    def add(self):
        result = self.num1 + self.num2
        return result
    
    def sub(self):
        result = self.num1 - self.num2
        return result
```

```python
sally = FourCal2(10,5)
```

```python
sally.add()
```
