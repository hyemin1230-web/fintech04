#### 1-50 사이의 숫자를 입력받아서 3의 배수면 3의 배수입니다.
- 3의 배우가 아니면 3의 배수가 아닙니다.
- 입력받기, 정수로 변환하기
- 1-50 사이의 숫자인가?
- 3의 배수인가? %3 == 0 
- 결과 출력

````python
num = int(input("정수로 숫자를 입력하세요: "))
print(num, type(num))
if 1<=num<=50:
    if num %3 == 0:
        print(f"{num}은 3의 배수입니다.")
    else: 
        print(f"{num}은 3의 배수가 아닙니다.")
else: 
    print("1-50 사이의 정수로 입력해주세요.")
````

````python
# 인덱스가 짝수 경우에는 건너뛰고 홀수인 경우에는 합쳐서 출력하세요.
# 단, 인덱스 20번까지만 출력하고 멈추세요.
for idx, item in enumerate(s):
    if idx %2 ==0:
        print("continue 앞에 있는 프린트문")
        continue
        print(idx)
        print("continue 뒤에 있는 프린트문")
    else: 
        if idx <= 20:
            print ((idx, item), end="")
        elif idx > 20:
               break
````
